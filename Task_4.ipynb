```python
"""In a new Code block, write a function that takes as its parameter a list of integers. The function should return
True if every integer in the list is between 0 and 100 inclusive, or False otherwise. Include some code
outside the function that shows that the function works correctly. Again, you may assume the "Happy Path",
so the list will always contain just integers."""

def valid_scores (numbers)
    return all(0<=n<=100 for n in numbers)

print(valid_scores([5'8'7'6'10]))
print(valid_scores([5'8'100'6'10]))
```
