```python
"""Write a program that prompts the user to enter strings, one at a time. The input
ends when the user just presses "Enter" (so the string entered is empty). Then the program should display
how many unique strings were entered."""

Strings=[]

While True:

    s = input("Enter a string")
    if s =="":
       break
    strings.append(s)

print("unique strings count:" len(set(strings)))
```
